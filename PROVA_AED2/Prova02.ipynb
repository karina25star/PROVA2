```python
#Célula 2

def media_consumo(consumos):
    soma = sum(consumos)
    quantidade = len(consumos)
    media = soma / quantidade
    return media
```

```python
#Célula 3

resultado = media_consumo(consumo_diario)
print("A média de consumo diário é:", resultado, "kWh")
```
